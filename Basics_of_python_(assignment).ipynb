***Exercise 1: Prime Numbers***

Write a Python program that checks whether a given number is prime or not. A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.


```python
# function to check the number is prime or not
def prime_number(n):
  count=0
  for i in range (2,n//2+1):
    if n%i==0:
      count+=1
    if count==1:
      return "not a prime number"
  if count==0:
    return "prime number"

# getting the input from the user
n=int(input("Enter the number:"))
if n>1:
  print(f"{n} is a {prime_number(n)}")   # calling the function
else:
  print("input should be greater than 1")
```

***Exercise 2: Product of Random Numbers***

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.


```python
# import a module to generate random number
import random as rd

# generate the random numbers
number1=rd.randint(1,10)
number2=rd.randint(1,10)

#asking the user to enter the product of random nubers
user_ans=int(input(f"Enter the product of {number1} and {number2} :"))

#checking the user input
if user_ans == number1*number2:
  print("Well done!!! The answer is correct...")
else:
  print("Oops!!! The answer is wrong...")


```

***Exercise 3: Squares of Even/Odd Numbers***

Create a Python script that prints the squares of all even or odd numbers within the range of 100 to 200. Choose either even or odd numbers and document your choice in the code.


```python
# initialize the list to store the output
output=[]

# function to get the square of even numbers
def even_square():
  for i in range (100,201,2):
    output.append(i*i)

# function to get the square of odd numbers
def odd_square():
  for i in range(100,200):
    if i%2!=0:
      output.append(i*i)

# getting the input from user (odd or even)
user_choice=input("Enter odd or even:")

# calling the function
if user_choice=='odd':
  odd_square()
else:
  even_square()

# printing the output
print(f"The {user_choice} squares are {output}")

```

***Exercise 4: Word counter***

write a program to count the number of words in a given text.
example:
input_text = "This is a sample text. This text will be used to demonstrate the word counter."
Expected output:
'This': 2
'is': 1
'a': 1
'sample': 1
'text.': 1


```python
# function for counting the word of the string
def word_counter(input_text):
  str=input_text.split(" ")
  str1=set()
  word_count={}
  for i in str:
    if i in str1:
      continue
    else:
      word_count[i]=str.count(i) # count() counts no. of words in the string
      str1.add(i)
  print(word_count)

# getting the input from the user
input_text=input("Enter the string:")

# calling the function
word_counter(input_text)


```

***Exercise 5: Check for Palindrome***

Write a Python function called is_palindrome that takes a string as input and returns True if the string is a palindrome, and False otherwise. A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward, ignoring spaces, punctuation, and capitalization.


```python
# function to check the string is palindrome
def is_palindrome(input_text):
  str=""
  for i in input_text:
    if i.isalnum():   # isalnum() checks the character is alphanumeric or not
      str+=i

  # reverse the string
  reversed_string=str[::-1]

  # compare the actual string and reversed string
  if str.lower()==reversed_string.lower():
    return True
  return False

# getting the input from the user
input_text=input("Enter the string:")
print(is_palindrome(input_text))
```
