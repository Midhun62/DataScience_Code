```python
#one hot encoding
genre_encoded=df['genre'].str.get_dummies(sep=',')
```

```python
type_encoded=pd.get_dummies(df['type'],dtype='int')
```

```python
#Min-Max scaler
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler()
df['rating']=pd.DataFrame(min_max.fit_transform(df[['rating']]))
```

```python
features=pd.concat([df['rating'],genre_encoded,type_encoded],axis=1)
```

Recommendation System:

Design a function to recommend anime based on cosine similarity.

Given a target anime, recommend a list of similar anime based on cosine similarity scores.

Experiment with different threshold values for similarity scores to adjust the recommendation list size.


```python
#cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
cos_similar=cosine_similarity(features)
```

```python
cos_similar
```

```python
name=df['name'].values
type(name)
```

```python
import numpy as np

def recommended_anime(anime_name,threshold):
  recommended_list=[]
  if anime_name in name:
    index=np.where(name==anime_name)[0][0]
    similar=sorted(list(enumerate(cos_similar[index])),reverse=True,key=lambda x:x[1])[1:threshold+1]
    print(f'Recommended anime of {anime_name}')
    print('-'*30)
    for anime in similar:
      recommended_list.append(name[anime[0]])
    return recommended_list
  else:
    print('Anime not found')
    return recommended_list
```

Evaluation:

Split the dataset into training and testing sets.

Evaluate the recommendation system using appropriate metrics such as precision, recall, and F1-score.

Analyze the performance of the recommendation system and identify areas of improvement.

```python
# splitting the data
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.3,random_state=100)
```

```python
# evaluation
from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_recommendation_system(test_set, top_n=10):
    y_true = []  # Actual similar anime (ground truth)
    y_pred = []  # Recommended anime

    for anime_name in test_set["name"]:
        if anime_name in name:
            true_similar = set(test_set[test_set["genre"] == test_set.loc[test_set["name"] == anime_name, "genre"].values[0]]["name"])
            # Get the predicted recommendations
            recommended_anime_list = set(recommended_anime(anime_name, top_n))
            y_true.append([1 if anime in true_similar else 0 for anime in recommended_anime_list])
            y_pred.append([1] * len(recommended_anime_list))
    y_true = [item for sublist in y_true for item in sublist]
    y_pred = [item for sublist in y_pred for item in sublist]
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-score: {f1:.2f}")
evaluate_recommendation_system(test)
```
