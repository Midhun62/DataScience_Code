*Provide a brief interpretation of these statistics.*

***Mean***

Mean is a measure of central tendency and also known as ***average***.It is defined as the sum of all observations divided by total number of observations.

Mean = Sum of all obsevations / Total no. of observations

In the dataset 'sales_data_with_discounts.csv' the mean is calculated for numerical columns such as volume(5.066667),Avg Price(10453.433333),Total Sales Value	(33812.835556),Discount Rate (%)	(15.155242),
Discount Amount	(3346.499424),Net Sales Value	(30466.336131) by using ***mean() of numpy*** library.




***Median***

Median is a measure of central tendency which refers to the ***middle value*** of the dataset.

In the dataset 'sales_data_with_discounts.csv' the median is calculated for numerical columns such as volume(4.000000),Avg Price(1450.000000),Total Sales Value (5700.000000),Discount Rate (%) (16.577766), Discount Amount (988.933733),Net Sales Value (4677.788059) by using ***median() of numpy*** library.

***Mode***

Mode is a measure of tendency which refers to the ***repeated value*** in the dataset i.e. the value which occurs most common in the dataset.

In the dataset 'sales_data_with_discounts.csv' the mode is calculated for numerical columns and categorical columns such as volume(3.0) and so on by using ***mode() of numpy*** library.

***Standard deviation***

Standard deviation is measure of dispersion which indicates the distance between the data points and the mean.

In the dataset 'sales_data_with_discounts.csv' the standard deviation is calculated for numerical columns  such as Volume	(4.231602),Avg Price	(18079.904840),Total Sales Value	(50535.074173),Discount Rate (%)	(4.220602),Discount Amount	(4509.902963),Net Sales Value	(46358.656624) by using ***std() of numpy*** library.

***Variance***

Variance is the measure of dispersion which describes how the data points are spread in the dataset (i.e. variance is the square of standard deviation).

In the dataset 'sales_data_with_discounts.csv' the variance is calculated for numerical columns by using ***var() of numpy*** library.

Standardization of Numerical Variables

●	Objective: To scale numerical variables for uniformity, improving the dataset’s suitability for analytical models.

●	Steps:

	Explain the concept of standardization (z-score normalization).
  
	Standardize the numerical columns using the formula: z=x-mu/sigma

	Show before and after comparisons of the data distributions.


Concept of standardization:



*  Standardization is the process of making all the numerical values in the same scale or in a standard form.
*   It is makes the model to learn the data in a easy manner.
*   It ranges from -3 to 3.






```python
#Standardize the numerical columns using the formula: z=x-mu/sigma

num_col.columns
```

```python
from sklearn.preprocessing import StandardScaler
sta_sca=StandardScaler()
standardized_d=sta_sca.fit_transform(num_col)
standardized_data=pd.DataFrame(standardized_d,columns=num_col.columns)
standardized_data
```

Show before and after comparisons of the data distributions.

```python
# before scaling
plt.hist(num_col)
plt.show()
```

```python
# After scaling
plt.hist(standardized_data)
plt.show()
```

From the histplot it is observed that the data distribution before scaling and after scaling is different.

Here,the standard deviation of data is 1 and the mean is 0.

Conversion of Categorical Data into Dummy Variables

●	Objective: To transform categorical variables into a format that can be provided to ML algorithms.

●	Steps:

	Discuss the need for converting categorical data into dummy variables (one-hot encoding).
	Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.
	Display a portion of the transformed dataset.


Discuss the need for converting categorical data into dummy variables (one-hot encoding).



*  One-hot encoding is one of the labeling technique in the exploratory data analysis process.

*   It is applied on categorical column especially on nomial category which cannot be ranked or ordered.

*   The need of converting categorical data into dummy variables is to make the ML model to understand the data since the models generally word on numerical columns.






Apply one-hot encoding to the categorical columns, creating binary (0 or 1) columns for each category.

```python
cat_col=df.select_dtypes('object')
cat_col.drop(columns='Date',inplace=True)
cat_col.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot=OneHotEncoder(drop='first',sparse_output=False)
one_hot_array=one_hot.fit_transform(cat_col)
one_hot_data=pd.DataFrame(one_hot_array,columns=one_hot.get_feature_names_out(cat_col.columns))
```
