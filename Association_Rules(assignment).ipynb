```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
enc=TransactionEncoder().fit(df_cleaned)
```

```python
one_hot=enc.transform(df_cleaned)
```

```python
data=pd.DataFrame(one_hot,columns=enc.columns_)
```

```python
data
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
# apriori algorithm
freq_patterns=apriori(data,min_support=0.01,use_colnames=True)
```

```python
# association rules
rules=association_rules(freq_patterns,min_threshold=0.01,metric='confidence')
```

```python
rules
```

```python
'''
lift:
--- measures the strength of rule compared to the random co-occurence of x and y.
--- lift=(support(x and y)/((support(x)*(support(y))))
--- helps to find the highly associated items which makes the better decisions for business.
'''
```

```python
'''
support:
--- provides fraction of transactions which contains the item x and y
--- support=support(x and y)/n


confidence:
--- Confidence measures how often item x appears in transactions that already contain item y.
--- confidence = support(x and y)/support(y)
'''
```
